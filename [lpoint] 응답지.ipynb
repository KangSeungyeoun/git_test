```python
ans = ans.dropna().reset_index()
```

```python
string_list = []
for i in ans.index :
    string = ans.loc[i,"reason"]
    string = string.replace("가격이 저렴하기 때문에","1_price")\
    .replace("문 앞으로 배송해주는 편리함 때문에","1_door")\
    .replace("다른 사람이 남긴 리뷰를 참고할 수 있기 때문에","1_review")\
    .replace("판매자와의 소통이 더 원활하기 때문에 (문의사항, 환불 등)","1_communication")\
    .replace("가격, 다양한 상품 등을 한 자리에서 비교해보고 살 수 있기 때문에","1_compare")\
    .replace("장바구니 담기, 찜하기, 간편결제 등의 온라인 구매 방식이 편리하기 떄문에","1_method")\
    .replace("주로 인터넷에서 구매한 제품이 더 만족스럽기 때문에 (신선도, 정품 확인 등)","1_satisfy")\
    .replace("시공간의 제약이 없기 때문에 (너무 늦거나 이른 시간의 쇼핑, 쇼핑몰 등이 멀어서 갈 수 없는 경우)","1_limit")\
    .replace(" ","")
    string_list.append(string)
```

```python
string2_list = []
for i in ans.index :
    string2 = ans.loc[i,"cause"]
    string2 = string2.replace("필요한 물품이 있을 시 필요에 따라 접속하여 구매한다.","2_need")\
    .replace("수시로 사이트에 접속하여 할인 정보를 탐색한 후 구매한다.","2_frequent")\
    .replace("인터넷이나 어플리케이션 광고 배너를 보고 접속하여 구매한다.","2_banner")\
    .replace("카톡이나 어플리케이션 PUSH 등을 통해 이벤트 등의 정보를 접하고 구매한다.","2_push")\
    .replace("구매 의사가 없던 상품의 리뷰 영상, 포스팅 등을 보고 구매 의사가 생겨 쇼핑몰에 접속하여 구매한다.","2_review")\
    .replace(" ","")
    string2_list.append(string2)
```

```python
string3_list = []
for i in ans.index :
    string3 = ans.loc[i,"item"]
    string3 = string3.replace("패션의류, 잡화, 뷰티","3_beauty")\
    .replace("식품","3_food")\
    .replace("생필품","3_daily")\
    .replace("홈데코, 문구, 취미, 반려동물","3_hobby")\
    .replace("컴퓨터, 디지털, 가전","3_digital")\
    .replace("스포츠, 건강, 렌탈","3_sports")\
    .replace("여행, 도서, 티켓, e쿠폰","3_culture")\
    .replace("그 외","3_else")\
    .replace(" ","")
    string3_list.append(string3)
```

```python
string2_list
```

```python
ans_list = ["purchase_time","1_price","1_door","1_review","1_communication","1_compare","1_method","1_satisfy",
            "1_limit","2_need","2_frequent","2_banner","2_push","2_review","3_beauty",
           "3_food","3_daily","3_hobby","3_digital","3_sports","3_culture","3_else"]
ans_df = pd.DataFrame(0, index=ans.index,columns=ans_list)
ans_df
```

```python
n = 0
for i in string_list :
    answers = i.split(",")
    for k in answers :
        ans_df.loc[n,k] += 1
    n += 1
```

```python
n = 0
for i in string2_list :
    answers = i.split(",")
    for k in answers :
        ans_df.loc[n,k] += 1
    n += 1
```

```python
n = 0
for i in string3_list :
    answers = i.split(",")
    for k in answers :
        ans_df.loc[n,k] += 1
    n += 1
```
